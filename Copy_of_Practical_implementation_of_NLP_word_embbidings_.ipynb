```python
from sklearn.metrics.pairwise import cosine_similarity
france=nlp.vocab["France"].vector
paris =nlp.vocab["Paris"].vector
berlin=nlp.vocab["Berlin"].vector
germany=nlp.vocab["Germany"].vector
result = france - paris + berlin
cosine_similarity([result],[germany])
#France-paris+berlin =Germany


```

```python
#cheak the imbalance in the dataset.
#if there is imbalance in the dataset then in that case the trining will be baised.
df['label'].value_counts()
```

```python
df['label_num'] = df['label'].map(
    {
    'FAKE':0,
    'REAL':1

 }
    )

```

```python
print(df.head())
```

```python
!python -m spacy download en_core_web_lg
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
#conert the text into vectors.
df['vector'] = None
for i in range(len(df)):
  nlp(df['text'][i]).vector
  df['vector'][i]=nlp(df['text'][i]).vector
```

```python
df.head(5)
```

```python
#Lets split it.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.vector.values,
                                                 df.label_num,
                                                 test_size = 0.2,
                                                 random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import numpy as np
X_train_stack = np.stack(X_train)
X_test_stack  = np.stack(X_test)
```

```python
print(X_train_stack.shape)
print(X_test_stack.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
X_train_stack_sc = scaler.fit_transform(X_train_stack)
X_test_stack_sc = scaler.transform(X_test_stack)



model.fit(X_train_stack_sc,y_train)
#every x_train vector row is converted into Stack and then passed into the model FIT.
```

```python
from sklearn.metrics import accuracy_score
y_pred =model.predict(X_test_stack_sc)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_pred)
```
